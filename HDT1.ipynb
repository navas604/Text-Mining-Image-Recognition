```python
# Función para escribir la imagen en disco y la lee de vuelta para poderla desplegar

def write_and_show(strImg, arrImg):
  cv2.imwrite(strImg, arrImg)
  imgTemp = cv2.imread(strImg)
  imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
  plt.imshow(imgTemp)
```

```python
# Función que recibe una imagen y un entero que determina el canal de color
# y devuelve la imagen con el canal de color activo especificado

def image_decompose_color(image, color):

  # Dimensiones de la imagen.
  alto = image.shape[0]
  ancho = image.shape[1]

  # Arreglo de salida para RGB
  imageSalida = np.zeros((alto, ancho, 3))

  for i in range(0, alto):
    for j in range(0, ancho):
      blue = image[i, j][2]
      green = image[i, j][1]
      red = image[i, j][0]

      if color == 1:  imageSalida[i, j] = [blue, 0, 0]
      if color == 2:  imageSalida[i, j] = [0, green, 0]
      if color == 3:  imageSalida[i, j] = [0, 0, red]
      if color == 10: imageSalida[i, j] = [0, green, red]
      if color == 20: imageSalida[i, j] = [blue, green, 0]
      if color == 30: imageSalida[i, j] = [blue, 0, red]

  write_and_show("Imagen_Salida.jpg", imageSalida)
```

```python
# Función que toma 3 imágenes en escala de grises en 3D, une las 3 imágenes y devuelve la imagen a color

def image_gray3D_to_color(image_gray_blue, image_gray_green, image_gray_red):

  # Dimensiones de la imagen.
  alto = image_gray_blue.shape[0]
  ancho = image_gray_blue.shape[1]

  # Arreglo de salida para RGB
  imageColor = np.zeros((alto, ancho, 3))

  for i in range(0, alto):
    for j in range(0, ancho):
      blue = image_gray_blue[i, j][2]
      green = image_gray_green[i, j][1]
      red = image_gray_red[i, j][0]

      imageColor[i, j] = [blue, green, red] 

  write_and_show("Imagen_Salida.jpg", imageColor)
```

```python
# Función que recibe una imagen y crea escala de grises en tres dimensiones, produciendo como salida 3 imágenes

def image_color_to_gray3D(image):

  # Dimensiones de la imagen.
  alto = image.shape[0]
  ancho = image.shape[1]

  # Arreglos para la salida
  gray_green = np.zeros((alto, ancho, 1))
  gray_blue = np.zeros((alto, ancho, 1))
  gray_red = np.zeros((alto, ancho, 1))

  # Lista en blanco para guardar las imagenes de salida
  images_list = []

  for i in range(0, alto):
    for j in range(0, ancho):
      gray_blue[i,j] = image[i,j][2]
      gray_green[i,j] = image[i,j][1]
      gray_red[i,j] = image[i,j][0] 

  # Escribir a disco las imagenes en escala de grises de salida
  cv2.imwrite("Imagen_Salida_gray_blue.jpg", gray_blue)
  cv2.imwrite("Imagen_Salida_gray_green.jpg", gray_green)
  cv2.imwrite("Imagen_Salida_gray_red.jpg", gray_red)

  # Leer las imagenes, convertir los colores de BGR a RGB y agregarla a la lista de imagenes
  imgTemp = cv2.imread("Imagen_Salida_gray_blue.jpg")
  images_list.append(cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB))

  imgTemp = cv2.imread("Imagen_Salida_gray_green.jpg")
  images_list.append(cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB))

  imgTemp = cv2.imread("Imagen_Salida_gray_red.jpg")
  images_list.append(cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB))

  colors = ["BLUE", "GREEN", "RED"]

  # Desplegar las 3 imagenes
  plt.figure(figsize=(20,20))
  for i in range(len(images_list)):
    plt.subplot(1, len(images_list), i + 1)
    plt.imshow(images_list[i])
    plt.title("GRAY {}".format(colors[i]), fontdict={"fontsize":16, "color":colors[i]})
  plt.show()
```

```python
image_color_to_gray3D(img)
```

```python
# Función que recibe una imagen y muestra el histograma de cada canal de color y el de escala de grises

def image_hist(image):

  # Dimensiones de la imagen.
  alto = image.shape[0]
  ancho = image.shape[1]

  # Arreglos para la salida
  green = np.zeros((alto, ancho, 1))
  blue = np.zeros((alto, ancho, 1))
  red = np.zeros((alto, ancho, 1))
  gray_aritmetica = np.zeros((alto, ancho, 1))

  # Lista en blanco para guardar las imagenes de salida
  images_list = []

  for i in range(0, alto):
    for j in range(0, ancho):
      blue[i,j] = image[i,j][2]
      green[i,j] = image[i,j][1]
      red[i,j] = image[i,j][0]
      gray_aritmetica[i,j] = np.mean(image[i,j])

  colors = ["BLUE", "GREEN", "RED", "GRAY"]

  # Agregar las imagenes a la lista
  images_list.append(blue.ravel())
  images_list.append(green.ravel())
  images_list.append(red.ravel())
  images_list.append(gray_aritmetica.ravel())

  # Desplegar las 3 imagenes
  plt.figure(figsize=(19,9))
  for i in range(len(images_list)):
    plt.subplot(2, 2, i+1)
    plt.hist(images_list[i], 256, [0,256], color=colors[i])
    plt.axvline(np.mean(images_list[i]), color="k", linestyle="--")
    plt.title("El valor esperado es: {}".format(np.round(np.mean(images_list[i]), 2)), fontdict={"fontsize":16, "color":colors[i]})
  plt.show()
```

```python
image_hist(img)
```

### ESCALA DE GRISES PONDERADA

La conversión de una imagen en color a escala de grises, es el equivalente a la luminancia de la imagen. El ojo humano es capaz de percibir distintas intesidades de luz en función del color que se observe, derivado de la respuesta del ojo al espectro de luz visible, por lo tanto para calcular el equivalente blanco y negro (escala de grises o luminancia) de la imagen se debe realizar como una media ponderada de las distintas componentes de color (RGB) de cada pixel.

Ecuación:

$$Y = 0.299R + 0.587G + 0.114B$$

Los factores de ponderación de cada componente de color nos indican la sensibilidad del ojo humano a las frecuencias del espectro cercanas al rojo, verde y azul (RGB). La visión humana es más sensible al verde, por ello tiene un coeficiente mayor y es menos sensible al azul, por ello tiene un coeficiente menor.



```python
# Cargar y mostrar imagen

img = cv2.imread("IMG_2461.JPG")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Función que recibe una imagen y muestra la imagen en escala de grises ponderado

def image_gray_ponderado(image):

  # Dimensiones de la imagen.
  alto = image.shape[0]
  ancho = image.shape[1]

  gray_ponderado = np.zeros((alto, ancho, 1))

  for i in range(0, alto):
    for j in range(0, ancho):
      pixel = image[i,j]
      
      gray_ponderado[i,j] = int(0.299*pixel[2] + 0.587*pixel[1] + 0.114*pixel[0])

  write_and_show("gray_ponderado.jpg", gray_ponderado)
```

```python
image_gray_ponderado(img)
```

### ESPACIO DE COLOR HSV

Es una representación tridimensional del color, definida en 1978 por Alvy Ray Smith, el cual se basa en 3 componentes: Matiz, Saturación y Valor (por sus siglas en ingles, HSV, Hue, Saturation, Value). También se le conoce con el nombre HSB (Hue, Saturation, Brightness) Matiz, Saturación y Brillo.

El modelo HSV es una transformación no lineal del modelo RGB en coordenadas cilíndricas.

El modelo RGB suele representarse como un cubo con coordenadas euclideanas, mientras que el modelo HSV suele representarse como un cono invertido y con coordenadas cilíndricas.

En el modelo HSV se agrupan las tonalidades de color, lo cual es una representación más parecida a la percepción humana de los colores y sus propiedades.

Las 3 componentes del modelo HSV son:
1.   **Matiz:** Es el ángulo que representa el matiz, definido cómo un ángulo entre 0º y 360º. En algunas ocasiones suele normalizarse a valores de 0 a 100%.
2.   **Saturación:** Es el nivel de saturación del color, definido por un valor entre 0 y 1, en donde 0 representa sin saturación (blanco) y 1 representa toda la intensidad, también suele representarse en porcentaje 0% - 100%. A menor saturación de un color, mayor tonalidad grisácea tendrá, por lo tanto se verá más decolorado.
3.   **Valor o Brillo:** Es el nivel de brillo del color, definido por un valor entre 0 y 1, en donde 0 es negro y 1 es blanco, también suele representarse en porcentaje 0% - 100%.
